## Linear Regression with higher-order terms

```python
df1 = df.copy()
df1 = df1.sort_values(by = 'x')                                #sorting values
X1 = df1[['x']]
Y1 = df1[['y']]
lm = linear_model.LinearRegression().fit(X1,Y1)
rl1 = lm.predict(X1)
nd = X1.to_numpy()
df                                                           #dataset for N=1  
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x='x',y='y',data=df,label='Datapoints')
plt.plot(nd,rl1,c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('N=1        UNDERFITTED MODEL')
plt.legend()
plt.show()
```

## LASSO  -  ROBUST MODEL

```python
poly = PolynomialFeatures(degree=10)
X_poly = poly.fit_transform(X1)
nd = X1.to_numpy()
from sklearn.linear_model import Lasso
l = [0.05,0.5,10,1000000]
for i in l:
    lasso = Lasso(alpha=i)
    lasso.fit(X_poly,Y1)
    y_pred = lasso.predict(X_poly)
    plt.figure(figsize=(10,5))
    plt.scatter(X1, Y1,label='Datapoints')
    plt.plot(nd, y_pred, c = 'red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'alpha={i}')
    plt.legend()
plt.show()
```
